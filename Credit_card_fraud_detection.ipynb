```python
import pandas as pa
import seaborn as sns
import numpy as ny
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
data = pa.read_csv('/content/drive/MyDrive/creditcard.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=data)
plt.title('Class Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.xlabel('Amount')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Time', y='Amount', data=data, hue='Class', alpha=0.5)
plt.title('Time vs Amount (Color-encoded by Class)')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
x = data.drop('Class', axis=1)
y = data['Class']

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
print(data['Class'].value_counts())

oversampler = RandomOverSampler(sampling_strategy=0.5)
x_resamp, y_resamp = oversampler.fit_resample(x, y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_resamp, y_resamp, test_size=0.2, random_state=2)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train.)
```

```python
data.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train_pred = model.predict(x_train)
x_train_pred
```

```python
accuracy = accuracy_score(y_train,x_train_pred)
accuracy
```

```python
x_test_pred = model.predict(x_test)
x_test_pred
```

```python
accuracy = accuracy_score(y_test,x_test_pred)
accuracy
```

```python
plt.scatter(y_test, x_test_pred)
plt.xlabel('Actual fraud')
plt.ylabel('Predicted fraud')
plt.title('Actual vs Predicted fraud')
plt.show()
```

```python
plt.bar(y_test, x_test_pred)
plt.xlabel('Actual fraud')
plt.ylabel('Predicted fraud')
plt.title('Actual vs Predicted fraud')
plt.show()
```
